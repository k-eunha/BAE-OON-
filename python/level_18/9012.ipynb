```python
num = int(input())
lst1 = []
lst2 = []

for i in range(num):
  lst1.append(input())
  count = 0
  for j in range(len(lst1[i])):
    check = lst1[i][j]
    if check == '(':
      count += 1
    else:
      count -= 1
      if count < 0:
        break
  if count == 0 :
    lst2.append("YES")
  else:
    lst2.append("NO")
for i in range(num):
  print(lst2[i])
```
